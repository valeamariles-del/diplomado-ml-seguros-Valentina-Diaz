## Parte 1 — Estructuras de Datos

### 1. Listas

Una **lista** es una secuencia **ordenada** y **mutable** de elementos.
Puede mezclar tipos. Equivale aproximadamente a `c()` en R.

| Propiedad | Valor |
|-----------|-------|
| Ordenada | Si — el orden se preserva |
| Mutable | Si — puedes agregar, eliminar y modificar |
| Duplicados | Si — permite valores repetidos |
| Sintaxis | `[elem1, elem2, ...]` |

````python
# ── Creacion ─────────────────────────────────────────────────────────────────
primas   = [2400, 6200, 1900, 14500, 3200]
nombres  = ['Ana', 'Luis', 'Maria', 'Carlos', 'Rosa']
mixta    = [1, 'GMM-001', True, 3.14, None]  # tipos distintos
vacia    = []

# Desde range
decada   = list(range(2015, 2026))  # [2015, 2016, ..., 2025]
print(decada)

print(type(primas))
print(f'Longitud: {len(primas)}')
````

````python
# ── Acceso por indice y slicing ──────────────────────────────────────────────
primas = [2400, 6200, 1900, 14500, 3200]

# Indices: 0-based, negativos desde el final
print(f'Primero:      {primas[0]}')    # 2400
print(f'Ultimo:       {primas[-1]}')   # 3200
print(f'Penultimo:    {primas[-2]}')   # 14500

# Slicing: [inicio:fin]  (fin excluido)
print(f'Primeros 3:   {primas[0:3]}')  # [2400, 6200, 1900]
print(f'Desde indice 2: {primas[2:]}') # [1900, 14500, 3200]
print(f'Cada 2:       {primas[::2]}')  # [2400, 1900, 3200]
print(f'Invertida:    {primas[::-1]}') # [3200, 14500, 1900, 6200, 2400]

# Verificar pertenencia
print(f'6200 en lista: {6200 in primas}')   # True
print(f'9999 en lista: {9999 in primas}')   # False
````

````python
# ── Metodos principales ──────────────────────────────────────────────────────
primas = [2400, 6200, 1900]

# Agregar
primas.append(8000)       # al final  → [2400, 6200, 1900, 8000]
primas.insert(0, 500)     # al inicio → [500, 2400, 6200, 1900, 8000]
primas.extend([3200, 4100])  # concatenar otra lista
print('Despues de agregar:', primas)

# Eliminar
primas.remove(500)        # por valor (primera ocurrencia)
ultimo = primas.pop()     # extrae y devuelve el ultimo
print(f'Extraido: {ultimo} | Lista: {primas}')

# Ordenar
primas.sort()                     # in-place (modifica la lista)
print('Ordenada:', primas)
nueva = sorted(primas, reverse=True)  # nueva lista descendente
print('Descendente:', nueva)

# Estadisticas
print(f'Sum:{sum(primas):,} | Min:{min(primas):,} | Max:{max(primas):,}')
print(f'Promedio: ${sum(primas)/len(primas):,.2f}')
````

````python
# ── Iterar con enumerate() y zip() ───────────────────────────────────────────
primas  = [2400, 6200, 1900, 14500, 3200]
nombres = ['Ana', 'Luis', 'Maria', 'Carlos', 'Rosa']
activas = [True, True, True, True, False]

# enumerate: indice + valor
print('Con enumerate:')
for i, prima in enumerate(primas):
    print(f'  Poliza {i+1}: ${prima:,.0f}')

# zip: dos listas en paralelo
print('Con zip:')
for nombre, prima, activa in zip(nombres, primas, activas):
    estado = 'ACTIVA' if activa else 'BAJA'
    print(f'  {nombre:<8} ${prima:>6,}  [{estado}]')
````

### 2. Tuplas

Una **tupla** es igual a una lista, pero **inmutable** — no puede modificarse
despues de crearse. Se usa para datos que no deben cambiar.

| Propiedad | Valor |
|-----------|-------|
| Ordenada | Si |
| Mutable | **No** — inmutable |
| Duplicados | Si |
| Sintaxis | `(elem1, elem2, ...)` |

````python
# ── Crear y usar tuplas ──────────────────────────────────────────────────────
rango_edad    = (18, 65)         # limites de edad asegurable
coordenadas   = (19.4326, -99.1332)  # CDMX
registro      = ('GMM-001', 3450.75, True)  # un registro fijo

# Acceso identico a lista
print(f'Edad min: {rango_edad[0]}, max: {rango_edad[1]}')
print(f'Latitud: {coordenadas[0]}')

# Desempaquetado (tuple unpacking)
edad_min, edad_max = rango_edad
lat, lon = coordenadas
print(f'Limites: {edad_min} - {edad_max}')

# Las tuplas SON inmutables
try:
    rango_edad[0] = 16   # esto FALLA
except TypeError as e:
    print(f'Error esperado: {e}')

# Uso tipico: funcion que devuelve multiples valores
def metricas_cartera(primas):
    return sum(primas), min(primas), max(primas), sum(primas)/len(primas)

total, minp, maxp, prom = metricas_cartera([2400, 6200, 1900, 14500])
print(f'Total:${total:,} Min:${minp:,} Max:${maxp:,} Prom:${prom:,.2f}')
````

````python
# ── Usos actuariales de tuplas ───────────────────────────────────────────────

# Tabla de tarifas como lista de tuplas (rango_edad, tasa)
tabla_tarifas = [
    ((18, 30), 0.018),
    ((31, 45), 0.022),
    ((46, 60), 0.031),
    ((61, 70), 0.048),
]

def obtener_tasa(edad, tabla):
    for (edad_min, edad_max), tasa in tabla:
        if edad_min <= edad <= edad_max:
            return tasa
    return None   # edad fuera de rango

# Probar con varias edades
edades_prueba = [25, 38, 52, 67, 75]
for edad in edades_prueba:
    tasa = obtener_tasa(edad, tabla_tarifas)
    if tasa:
        print(f'Edad {edad}: tasa = {tasa:.1%}')
    else:
        print(f'Edad {edad}: fuera de rango asegurable')
````

### 3. Diccionarios

Un **diccionario** almacena pares **clave: valor**. Es la estructura mas importante
para datos estructurados — una poliza, una fila de pandas, una respuesta de API.

| Propiedad | Valor |
|-----------|-------|
| Ordenada | Si (Python 3.7+) |
| Mutable | Si |
| Duplicados | Llaves unicas (valores pueden repetirse) |
| Sintaxis | `{clave: valor, ...}` |

````python
# ── Crear y acceder ──────────────────────────────────────────────────────────
poliza = {
    'id':         'GMM-2026-001',
    'titular':    'Maria Garcia',
    'edad':       32,
    'prima':      2400.0,
    'activa':     True,
    'coberturas': ['Hospitalizacion', 'Cirugia', 'Dental'],
}

# Acceso por clave
print(poliza['titular'])       # Maria Garcia
print(poliza['prima'])         # 2400.0
print(poliza['coberturas'][0]) # Hospitalizacion

# .get() — valor por defecto si la clave no existe
print(poliza.get('deducible', 5000))   # 5000  (no esta en el dict)
print(poliza.get('activa', False))     # True  (si esta)

# Verificar si existe una clave
print('edad' in poliza)        # True
print('imc'  in poliza)        # False
print(len(poliza))             # 6 pares clave-valor
````

````python
# ── Modificar diccionarios ───────────────────────────────────────────────────
poliza = {'id': 'GMM-001', 'prima': 2400, 'activa': True}

# Agregar nueva clave
poliza['deducible'] = 5000
poliza['anio_emision'] = 2026

# Modificar valor existente
poliza['prima'] = 2600   # actualizacion de prima

# Eliminar clave
del poliza['activa']
valor = poliza.pop('anio_emision', None)  # extrae y devuelve
print(f'Extraido: {valor}')
print(poliza)

# Recorrer el diccionario
print('Llaves:', list(poliza.keys()))
print('Valores:', list(poliza.values()))
print()
for clave, valor in poliza.items():
    print(f'  {clave:<15}: {valor}')
````

````python
# ── Patron fundamental: lista de diccionarios = tabla de datos ───────────────
# Esto es EXACTAMENTE como pandas almacena internamente un DataFrame

cartera = [
    {'id':'GMM-001', 'titular':'Ana',    'edad':28, 'prima':2400, 'siniest':0},
    {'id':'GMM-002', 'titular':'Luis',   'edad':45, 'prima':6200, 'siniest':1},
    {'id':'GMM-003', 'titular':'Maria',  'edad':32, 'prima':1900, 'siniest':0},
    {'id':'GMM-004', 'titular':'Carlos', 'edad':62, 'prima':14500,'siniest':3},
    {'id':'GMM-005', 'titular':'Rosa',   'edad':38, 'prima':3200, 'siniest':0},
]

# Acceder a un campo de todas las polizas
todas_primas = [p['prima'] for p in cartera]
print(f'Todas las primas: {todas_primas}')
print(f'Total: ${sum(todas_primas):,}')

# Filtrar polizas con siniestros
con_siniest = [p for p in cartera if p['siniest'] > 0]
print(f'Con siniestros: {len(con_siniest)} polizas')
for p in con_siniest:
    print(f"  {p['titular']}: {p['siniest']} siniestro(s)")

# Tabla de tarifas y calcular prima
tarifas_edad = {'18-30':0.018, '31-45':0.022, '46-60':0.031, '61+':0.048}

def grupo_edad(edad):
    if edad <= 30: return '18-30'
    elif edad <= 45: return '31-45'
    elif edad <= 60: return '46-60'
    return '61+'

for p in cartera:
    g = grupo_edad(p['edad'])
    tasa = tarifas_edad[g]
    prima_calc = 500_000 * tasa
    print(f"{p['titular']:<8} edad:{p['edad']:>3} grupo:{g} prima_calc:${prima_calc:,.0f}")
````

### 4. Conjuntos (Sets)

Un **set** es una coleccion **sin duplicados** y **sin orden**.
Ideal para encontrar valores unicos y operar conjuntos (union, interseccion).

| Propiedad | Valor |
|-----------|-------|
| Ordenada | **No** |
| Mutable | Si |
| Duplicados | **No** — los elimina automaticamente |
| Sintaxis | `{elem1, elem2, ...}` o `set(lista)` |

````python
# ── Crear y operar sets ──────────────────────────────────────────────────────

# Elimina duplicados automaticamente
ramos_cartera = ['GMM', 'Autos', 'GMM', 'Vida', 'Autos', 'GMM', 'Vida']
ramos_unicos  = set(ramos_cartera)
print(f'Ramos distintos: {ramos_unicos}')
print(f'Cantidad: {len(ramos_unicos)}')

# Coberturas de dos productos distintos
coberturas_basico  = {'Hospitalizacion', 'Cirugia', 'Urgencias'}
coberturas_premium = {'Hospitalizacion', 'Cirugia', 'Urgencias',
                      'Maternidad', 'Dental', 'Vision'}

# Union: todo lo que tiene basico O premium
print(f'Union:        {coberturas_basico | coberturas_premium}')

# Interseccion: solo lo que tienen AMBOS
print(f'Interseccion: {coberturas_basico & coberturas_premium}')

# Diferencia: en premium pero NO en basico
solo_premium = coberturas_premium - coberturas_basico
print(f'Solo en premium: {solo_premium}')

# Verificar subconjunto
print(f'Basico es subconjunto de premium: {coberturas_basico <= coberturas_premium}')
````

````python
# ── Aplicaciones practicas de sets en seguros ────────────────────────────────

# Encontrar ramos que tiene un agente pero no otro
ramos_agente_A = {'GMM', 'Autos', 'Vida', 'Accidentes', 'Viajes'}
ramos_agente_B = {'GMM', 'Vida', 'Hogar', 'Negocio'}

exclusivos_A = ramos_agente_A - ramos_agente_B
exclusivos_B = ramos_agente_B - ramos_agente_A
comunes      = ramos_agente_A & ramos_agente_B
todos        = ramos_agente_A | ramos_agente_B

print(f'Solo agente A:  {exclusivos_A}')
print(f'Solo agente B:  {exclusivos_B}')
print(f'Ambos venden:   {comunes}')
print(f'Cartera total:  {todos}')

# Detectar polizas duplicadas (mismo id)
ids_sistema_A = ['GMM-001', 'GMM-002', 'GMM-003', 'GMM-004']
ids_sistema_B = ['GMM-003', 'GMM-004', 'GMM-005', 'GMM-006']

duplicados = set(ids_sistema_A) & set(ids_sistema_B)
print(f'Polizas en ambos sistemas: {duplicados}')
````

### 5. if / elif / else

La **indentacion** (4 espacios) define los bloques de codigo en Python.
No hay llaves `{}` — el espacio en blanco tiene significado.

```
if condicion:
    # bloque si condicion es True
elif otra_condicion:
    # bloque si otra_condicion es True
else:
    # bloque si ninguna condicion fue True
```

````python
# ── if basico ────────────────────────────────────────────────────────────────
prima = 15000

# if simple
if prima > 10000:
    print('Prima alta — requiere autorizacion especial')

# if-else
if prima >= 5000:
    descuento = 0.05
else:
    descuento = 0.0
print(f'Descuento: {descuento:.0%}')

# if-elif-else: clasificacion por nivel de prima
if prima < 2000:
    nivel = 'Basico'
elif prima < 5000:
    nivel = 'Estandar'
elif prima < 10000:
    nivel = 'Preferente'
else:
    nivel = 'Premium'
print(f'Nivel de cliente: {nivel}')
````

````python
# ── Reglas de underwriting con if ────────────────────────────────────────────

def evaluar_solicitud(edad, imc, preexistencias, siniestros_3a, suma_aseg):
    '''Evalua si una solicitud de GMM puede aceptarse automaticamente.'''

    # Rechazos directos
    if edad < 18 or edad > 70:
        return 'DECLINA', 'Edad fuera de rango asegurable'
    if imc >= 40:
        return 'DECLINA', 'IMC >= 40'
    if siniestros_3a >= 4:
        return 'DECLINA', 'Mas de 3 siniestros en 3 anios'

    # Revision medica requerida
    if edad >= 55 and suma_aseg > 2_000_000:
        return 'REVISION', 'Edad >= 55 con suma alta'
    if imc >= 30 and preexistencias:
        return 'REVISION', 'IMC alto + preexistencias'
    if siniestros_3a >= 2:
        return 'REVISION', '2 o mas siniestros recientes'

    return 'ACEPTA', 'Cumple todos los criterios'


solicitudes = [
    ('Ana Torres',   28, 22.5, False, 0, 500_000),
    ('Luis Perez',   58, 28.1, True,  1, 3_000_000),
    ('Rosa Mendez',  35, 38.5, False, 0, 300_000),
    ('Jorge Silva',  72, 25.0, False, 0, 500_000),
    ('Elena Castro', 45, 31.2, True,  3, 800_000),
]

print(f'{"Nombre":<16} {"Resultado":<12} {"Razon"}')
print('-' * 60)
for nombre, edad, imc, preex, sin, suma in solicitudes:
    resultado, razon = evaluar_solicitud(edad, imc, preex, sin, suma)
    print(f'{nombre:<16} {resultado:<12} {razon}')
````

### 6. for — Iterar Sobre Colecciones

El `for` en Python itera sobre **cualquier cosa iterable**: listas, tuplas,
diccionarios, strings, rangos...

````python
# ── Formas del for ───────────────────────────────────────────────────────────
primas = [2400, 6200, 1900, 14500, 3200]

# Forma basica
total = 0
for prima in primas:
    total += prima
print(f'Total (for): ${total:,}  |  Total (sum): ${sum(primas):,}')

# Con enumerate: indice + valor
for i, prima in enumerate(primas, 1):  # empieza en 1
    print(f'  Poliza {i}: ${prima:,.0f}')
````

````python
# ── zip, range y break/continue ─────────────────────────────────────────────
nombres = ['Ana', 'Luis', 'Maria', 'Carlos', 'Rosa']
primas  = [2400, 6200, 1900, 14500, 3200]
activas = [True, True, True, True, False]

# zip: iterar multiples listas en paralelo
print('Polizas activas:')
for nom, prima, activa in zip(nombres, primas, activas):
    if not activa:
        continue   # salta las inactivas
    print(f'  {nom:<8} ${prima:,.0f}')

# break: salir antes de terminar
print('Primera prima > 5000:')
for i, prima in enumerate(primas):
    if prima > 5000:
        print(f'  Poliza {i+1}: ${prima:,}')
        break

# Iterar sobre range
print('Meses del anio:')
for mes in range(1, 13):
    print(f'  Mes {mes:02d}', end=' | ' if mes % 4 != 0 else '\n')
````

````python
# ── Iterar diccionarios ──────────────────────────────────────────────────────
cartera = [
    {'id':'GMM-001', 'titular':'Ana',   'prima':2400, 'siniest':0},
    {'id':'GMM-002', 'titular':'Luis',  'prima':6200, 'siniest':1},
    {'id':'GMM-003', 'titular':'Maria', 'prima':1900, 'siniest':0},
]

# Calcular metricas sobre la cartera
total_prima = 0
total_siniest = 0

for poliza in cartera:
    total_prima   += poliza['prima']
    total_siniest += poliza['siniest']
    # Clasificacion por prima
    nivel = 'Alto' if poliza['prima'] >= 5000 else 'Bajo'
    print(f"{poliza['id']}  {poliza['titular']:<8}  ${poliza['prima']:>6,}  [{nivel}]")

print(f'Total prima: ${total_prima:,}')
print(f'Total siniestros: {total_siniest}')
print(f'Loss ratio implícito (asumiendo 50k/siniest): {total_siniest*50000/total_prima:.1%}')
````

### 7. List Comprehensions

Las comprehensions son la forma **Pytonica** de crear listas a partir de iterables.
Son mas concisas y generalmente mas rapidas que un for loop equivalente.

**Sintaxis:** `[expresion for elemento in iterable if condicion]`

````python
# ── List comprehensions ──────────────────────────────────────────────────────
primas  = [2400, 6200, 1900, 14500, 3200, 8800, 1850, 7300]
activas = [True, True, True, True,  False,True, True, False]

# Basica: transformar cada elemento
primas_mensual = [p / 12 for p in primas]
print('Primas mensuales:', [round(m, 2) for m in primas_mensual])

# Con condicion: filtrar
primas_altas = [p for p in primas if p > 5000]
print(f'Primas > $5,000: {primas_altas}')

# Transformar + filtrar
primas_activas = [p for p, a in zip(primas, activas) if a]
print(f'Solo activas: {primas_activas}')

# Clasificacion con ternario
niveles = ['Alto' if p >= 8000 else 'Medio' if p >= 3000 else 'Bajo'
           for p in primas]
print(f'Niveles: {niveles}')

# Dict comprehension
ids = [f'GMM-{i+1:03d}' for i in range(len(primas))]
cartera_dict = {id_: prima for id_, prima in zip(ids, primas)}
print('Cartera dict:', cartera_dict)
````

````python
# ── Comprehensions avanzadas ─────────────────────────────────────────────────
cartera = [
    {'id':'GMM-001', 'titular':'Ana',   'edad':28, 'prima':2400, 'ramo':'GMM'},
    {'id':'AUT-001', 'titular':'Luis',  'edad':45, 'prima':5800, 'ramo':'Autos'},
    {'id':'GMM-002', 'titular':'Maria', 'edad':32, 'prima':1900, 'ramo':'GMM'},
    {'id':'VID-001', 'titular':'Carlos','edad':62, 'prima':9200, 'ramo':'Vida'},
    {'id':'AUT-002', 'titular':'Rosa',  'edad':38, 'prima':4100, 'ramo':'Autos'},
]

# Filtrar por ramo
polizas_gmm   = [p for p in cartera if p['ramo'] == 'GMM']
polizas_autos = [p for p in cartera if p['ramo'] == 'Autos']
print(f'GMM: {len(polizas_gmm)} polizas | Autos: {len(polizas_autos)} polizas')

# Extraer un campo de todas las polizas de un ramo
primas_gmm = [p['prima'] for p in cartera if p['ramo'] == 'GMM']
print(f'Primas GMM: {primas_gmm} | Promedio: ${sum(primas_gmm)/len(primas_gmm):,.2f}')

# Ramos distintos en la cartera (usar set + comprehension)
ramos_unicos = {p['ramo'] for p in cartera}  # set comprehension
print(f'Ramos en cartera: {ramos_unicos}')

# Resumen por ramo con dict comprehension
resumen = {
    ramo: sum(p['prima'] for p in cartera if p['ramo'] == ramo)
    for ramo in ramos_unicos
}
print('Prima total por ramo:')
for ramo, total in sorted(resumen.items(), key=lambda x: x[1], reverse=True):
    print(f'  {ramo:<8}: ${total:,.0f}')
````

### 8. while — Repetir Mientras se Cumpla una Condicion

El `while` repite un bloque **mientras** una condicion sea `True`.
Siempre asegurate de que la condicion eventualmente se vuelva `False`
o usa `break` para salir — de lo contrario sera un loop infinito.

````python
# ── while basico ─────────────────────────────────────────────────────────────
# Patron 1: contador
contador = 1
while contador <= 5:
    print(f'Iteracion {contador}')
    contador += 1   # FUNDAMENTAL: modificar la condicion

# Patron 2: acumulador
reserva_requerida = 500_000
reserva_actual    = 0
pago_mensual      = 45_000
mes = 0

while reserva_actual < reserva_requerida:
    reserva_actual += pago_mensual
    mes += 1

print(f'Meses para cubrir reserva: {mes}')
print(f'Reserva final: ${reserva_actual:,.0f}')
````

````python
# ── while con break y continue ───────────────────────────────────────────────
import random
random.seed(42)

# Simular revision de polizas hasta encontrar una de alto riesgo
polizas_revisadas = 0
while True:
    # Simular una poliza con IMC aleatorio
    imc = random.uniform(18, 45)
    polizas_revisadas += 1

    if imc < 30:
        continue   # IMC normal, siguiente poliza

    # IMC >= 30: registrar y seguir
    print(f'Poliza {polizas_revisadas}: IMC = {imc:.1f} — revision medica')

    if polizas_revisadas >= 10:
        break  # limite de revision

print(f'Total revisadas: {polizas_revisadas}')
````

<a id='9-try'></a>
### 9. try / except — Manejar Errores

En produccion, los datos nunca son perfectos. El `try/except` permite
manejar errores de forma elegante sin que el programa se detenga.

````python
# ── Errores comunes en datos de seguros ──────────────────────────────────────

# Sin manejo de errores: el programa se detiene
# prima = 5000 / 0     ← ZeroDivisionError
# edad = int('N/A')    ← ValueError
# d = {}; d['prima']   ← KeyError

# ── Con try/except ────────────────────────────────────────────────────────────
registros_csv = [
    {'id': 'GMM-001', 'prima': '2400',  'exposicion': '365'},
    {'id': 'GMM-002', 'prima': 'N/A',   'exposicion': '180'},   # prima sucia
    {'id': 'GMM-003', 'prima': '1900',  'exposicion': '0'},     # div por cero
    {'id': 'GMM-004', 'prima': '14500', 'exposicion': '365'},
]

print(f'{"ID":<10} {"Prima":>8} {"Exp":>5} {"Prima/dia":>12} {"Estado"}')
print('-' * 55)

for reg in registros_csv:
    estado = 'OK'
    prima_dia = None
    try:
        prima = float(reg['prima'])       # puede ser ValueError
        exp   = int(reg['exposicion'])
        prima_dia = prima / exp            # puede ser ZeroDivisionError
    except ValueError:
        estado = 'ERROR: prima no numerica'
    except ZeroDivisionError:
        estado = 'ERROR: exposicion cero'
    except KeyError as e:
        estado = f'ERROR: campo {e} faltante'
    finally:
        pd_str = f'${prima_dia:,.4f}' if prima_dia else 'N/A'
        print(f"{reg['id']:<10} {reg['prima']:>8} {reg['exposicion']:>5} {pd_str:>12} {estado}")
````

### 10. Lambda, map y filter

Las funciones **lambda** son funciones anonimas de una sola linea.
Muy usadas con `sorted()`, `map()`, `filter()` y con pandas (`.apply()`).

````python
# ── Lambda basica ────────────────────────────────────────────────────────────
# def normal vs lambda
def prima_mensual_normal(prima_anual):
    return prima_anual / 12

prima_mensual_lambda = lambda p: p / 12

# Equivalentes
print(prima_mensual_normal(3600))  # 300.0
print(prima_mensual_lambda(3600))  # 300.0

# Lambda con multiple argumentos
prima_con_factor = lambda prima, factor=1.16: prima * factor
print(f'Con IVA 16%: ${prima_con_factor(3450):,.2f}')
print(f'Con recargo 20%: ${prima_con_factor(3450, 1.20):,.2f}')
````

````python
# ── sorted() con clave lambda ────────────────────────────────────────────────
cartera = [
    {'id':'GMM-001', 'titular':'Ana',   'prima':2400, 'edad':28},
    {'id':'GMM-002', 'titular':'Luis',  'prima':6200, 'edad':45},
    {'id':'GMM-003', 'titular':'Maria', 'prima':1900, 'edad':32},
    {'id':'GMM-004', 'titular':'Carlos','prima':14500,'edad':62},
    {'id':'GMM-005', 'titular':'Rosa',  'prima':3200, 'edad':38},
]

# Ordenar por prima descendente
por_prima = sorted(cartera, key=lambda p: p['prima'], reverse=True)
print('Ranking por prima:')
for i, p in enumerate(por_prima, 1):
    print(f'  {i}. {p["titular"]:<8} ${p["prima"]:>6,}')

# Ordenar por edad
por_edad = sorted(cartera, key=lambda p: p['edad'])
print('Por edad:')
for p in por_edad:
    print(f'  {p["titular"]:<8} edad:{p["edad"]}')
````

````python
# ── map() y filter() ─────────────────────────────────────────────────────────
primas = [2400, 6200, 1900, 14500, 3200, 8800]

# map(): aplicar funcion a cada elemento
primas_con_iva = list(map(lambda p: p * 1.16, primas))
print('Con IVA:', [f'${p:,.2f}' for p in primas_con_iva])

# filter(): filtrar elementos
primas_altas = list(filter(lambda p: p > 5000, primas))
print(f'Primas > 5000: {primas_altas}')

# Equivalente con comprehension (mas Pythonico):
primas_altas_comp = [p for p in primas if p > 5000]
print(f'Equivalente:   {primas_altas_comp}')

# map + filter combinados
primas_altas_con_iva = list(map(
    lambda p: p * 1.16,
    filter(lambda p: p > 5000, primas)
))
print('Altas con IVA:', [f'${p:,.2f}' for p in primas_altas_con_iva])
````

## 11. Ejercicio Integrador de Cierre

**Contexto:** Eres el analista de datos del area de Autos de una aseguradora.
Tienes una cartera de 5 polizas y una tabla de tarifas por tipo de vehiculo.
Debes generar el reporte mensual del area.

````python
# ── Datos ────────────────────────────────────────────────────────────────────
cartera = [
    {'id':'A-101','nombre':'Sofia R.', 'edad':24,'vehiculo':'Sedan',    'siniest':0,'suma':300_000},
    {'id':'A-102','nombre':'Miguel T.','edad':38,'vehiculo':'SUV',      'siniest':1,'suma':500_000},
    {'id':'A-103','nombre':'Laura G.', 'edad':31,'vehiculo':'Hatchback','siniest':0,'suma':250_000},
    {'id':'A-104','nombre':'Roberto M.','edad':55,'vehiculo':'SUV',     'siniest':2,'suma':600_000},
    {'id':'A-105','nombre':'Carmen P.','edad':42,'vehiculo':'Sedan',    'siniest':0,'suma':350_000},
]
tarifas = {'Sedan':0.025, 'SUV':0.035, 'Hatchback':0.022, 'Pick-up':0.040}
print(f'{len(cartera)} polizas cargadas')
````

````python
# ── Tarea 1: Calcular prima y clasificar riesgo ───────────────────────────────
def clasificar_riesgo(siniest):
    if siniest >= 2: return 'ALTO'
    elif siniest == 1: return 'MEDIO'
    return 'BAJO'

for p in cartera:
    p['prima'] = p['suma'] * tarifas[p['vehiculo']]
    p['riesgo'] = clasificar_riesgo(p['siniest'])

print('Calculo de primas y clasificacion:')
for p in cartera:
    print(f"  {p['id']} {p['nombre']:<12} ${p['prima']:>8,.0f}  Riesgo:{p['riesgo']}")
````

````python
# ── Tarea 2: Reporte tabular completo ────────────────────────────────────────
print('=' * 65)
print('  REPORTE MENSUAL — CARTERA AUTOS')
print('=' * 65)
print(f'{"ID":<8} {"Nombre":<14} {"Vehiculo":<12} {"Prima":>9} {"Riesgo":<8} {"Sinist"}')
print('-' * 65)
for p in cartera:
    print(f"{p['id']:<8} {p['nombre']:<14} {p['vehiculo']:<12} ",
          f"${p['prima']:>8,.0f} {p['riesgo']:<8} {p['siniest']}")
print('-' * 65)

prima_total = sum(p['prima'] for p in cartera)
total_siniest = sum(p['siniest'] for p in cartera)
print(f'{'TOTAL':<36} ${prima_total:>8,.0f}')
print(f'Siniestros totales: {total_siniest}')
````

````python
# ── Tarea 3: Analisis con comprehensions ─────────────────────────────────────

# Primas de SUVs
primas_suv = [p['prima'] for p in cartera if p['vehiculo'] == 'SUV']
print(f'Primas de SUVs: {primas_suv}')
print(f'Prima promedio SUV: ${sum(primas_suv)/len(primas_suv):,.2f}')

# Polizas de alto riesgo
alto_riesgo = [p['nombre'] for p in cartera if p['riesgo'] == 'ALTO']
print(f'Alto riesgo: {alto_riesgo}')

# Vehiculos distintos en la cartera
vehiculos_unicos = {p['vehiculo'] for p in cartera}
print(f'Vehiculos en cartera: {vehiculos_unicos}')

# Prima promedio general
prima_prom = sum(p['prima'] for p in cartera) / len(cartera)
print(f'Prima promedio general: ${prima_prom:,.2f}')
````

````python
# ── Tarea 4: Ranking por prima (bonus) ───────────────────────────────────────
ranking = sorted(cartera, key=lambda p: p['prima'], reverse=True)

print('RANKING POR PRIMA (mayor a menor):')
print(f'{"Pos":<5} {"Nombre":<14} {"Vehiculo":<12} {"Prima":>9}')
print('-' * 45)
for i, p in enumerate(ranking, 1):
    print(f'{i:<5} {p["nombre"]:<14} {p["vehiculo"]:<12} ${p["prima"]:>8,.0f}')

# Resumen por vehiculo con dict comprehension
print()
print('PRIMAS POR TIPO DE VEHICULO:')
veh_tipos = {p['vehiculo'] for p in cartera}
resumen = {
    veh: [p['prima'] for p in cartera if p['vehiculo'] == veh]
    for veh in veh_tipos
}
for veh, ps in sorted(resumen.items()):
    print(f'  {veh:<12} N={len(ps)} | Total: ${sum(ps):>8,.0f} | Prom: ${sum(ps)/len(ps):>8,.0f}')
````
