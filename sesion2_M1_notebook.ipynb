### 3. Flujo: add → commit → push

```
[Editas archivos]  →  git add  →  git commit  →  git push
Working Directory     Staging      Repository     GitHub
```

| Paso | Comando | Que hace |
|------|---------|----------|
| 1 | `git status` | Ver que cambio |
| 2 | `git add .` | Preparar todos los cambios |
| 3 | `git commit -m 'msg'` | Guardar punto de control |
| 4 | `git push` | Subir a GitHub |

#### Autenticacion con GitHub

La primera vez que hagas `git push`, GitHub pedira autenticacion.

1. GitHub → foto de perfil → **Settings**
2. **Developer settings** → **Personal access tokens** → **Tokens (classic)**

4. **Copia el token** (solo se muestra una vez)
5. Cuando Git pida contrasena, pega el token

## Parte 2 — Sintaxis Basica de Python (T2)

### 4. Tipos de Datos — Panorama

En Python **todo tiene un tipo**. La funcion `type()` te dice el tipo de cualquier valor.

| Tipo | Clase Python | Ejemplo actuarial |
|------|-------------|-------------------|
| Entero | `int` | numero de siniestros, year, edad |
| Decimal | `float` | prima, tasa, frecuencia |
| Texto | `str` | nombre asegurado, numero de poliza, ramo |
| Logico | `bool` | poliza activa, tiene siniestros |
| Sin valor | `NoneType` | beneficiario no registrado, campo faltante |
| Fecha | `datetime.date` | fecha emision, vencimiento, siniestro |

````python
# La funcion type() revela el tipo de cualquier valor
print(type(35))               # <class 'int'>
print(type(3450.75))          # <class 'float'>
print(type('GMM-2026-001'))   # <class 'str'>
print(type(True))             # <class 'bool'>
print(type(None))             # <class 'NoneType'>

# Python infiere el tipo automaticamente
x = 10       # int
y = 10.0     # float  (el punto decimal lo cambia todo)
print(type(x), type(y))
````

### 5. int y float — Numeros

````python
# ── int: numeros enteros ─────────────────────────────────────────────────────
edad          = 45
num_siniestros = 3
anio_emision  = 2026
total_polizas = 8_500   # _ es solo separador visual, Python lo ignora

print(f'Edad:          {edad}')
print(f'Siniestros:    {num_siniestros}')
print(f'Total polizas: {total_polizas:,}')   # :, agrega separador de miles
print(f'Tipo: {type(total_polizas)}')

# Operadores enteros especificos
meses_vigencia = 350
meses_completos = meses_vigencia // 30   # division entera
dias_sobrantes  = meses_vigencia % 30    # modulo (residuo)
print(f'Meses completos: {meses_completos}, dias sobrantes: {dias_sobrantes}')
````

````python
# ── float: numeros decimales ─────────────────────────────────────────────────
prima_pura      = 18_000.0
tasa_siniestral = 0.783      # 78.3%
factor_inflacion = 1.065

print(f'Prima pura:       ${prima_pura:>14,.2f}')
print(f'Tasa siniestral:  {tasa_siniestral:>14.1%}')
print(f'Factor inflacion: {factor_inflacion:>14.4f}')

# Precision decimal — importante en actuaria
resultado = 0.1 + 0.2
print(f'0.1 + 0.2 = {resultado}')           # No es exactamente 0.3
print(f'round:     {round(resultado, 2)}')  # Solucion: round()
````

````python
# ── Funciones matematicas ─────────────────────────────────────────────────────
import math

# Raiz cuadrada — volatilidad, desvio estandar
vol_diaria = 0.015
vol_anual  = vol_diaria * math.sqrt(252)   # 252 dias habiles
print(f'Volatilidad anual: {vol_anual:.4f}')

# Logaritmo natural — modelos GLM con link log
suma_asegurada = 1_000_000
print(f'ln(1,000,000) = {math.log(suma_asegurada):.4f}')

# Exponencial — reversa del logaritmo
coef_edad  = 0.032   # coeficiente de edad en un GLM
multiplicador = math.exp(coef_edad)
print(f'Efecto por year de edad: {multiplicador:.4f}')
print(f'10 years mayor paga {multiplicador**10:.2f}x mas')
````

````python
# ── Calculo actuarial completo — prima comercial GMM ─────────────────────────
suma_asegurada    = 500_000
tasa_riesgo       = 0.022
recargo_seguridad = 0.15
recargo_gastos    = 0.08
comision_agente   = 0.10
iva               = 1.16

prima_pura      = suma_asegurada * tasa_riesgo
prima_neta      = prima_pura * (1 + recargo_seguridad)
prima_comercial = prima_neta / (1 - recargo_gastos - comision_agente)
prima_total     = prima_comercial * iva
prima_mensual   = prima_total / 12

print('-' * 50)
print('  ESTRUCTURA DE PRIMA — GMM Individual')
print('-' * 50)
print(f'  Suma asegurada    : ${suma_asegurada:>12,.2f}')
print(f'  Prima pura        : ${prima_pura:>12,.2f}')
print(f'  Prima neta        : ${prima_neta:>12,.2f}')
print(f'  Prima comercial   : ${prima_comercial:>12,.2f}')
print(f'  Prima total + IVA : ${prima_total:>12,.2f}')
print(f'  Prima mensual     : ${prima_mensual:>12,.2f}')
print('-' * 50)
````

### 6. str — Cadenas de Texto

Las cadenas son omnipresentes en seguros: nombres de asegurados, numeros de poliza,
ramos, coberturas. Tambien son lo primero que aparece al leer un CSV.

````python
# ── Creacion y acceso por indice ─────────────────────────────────────────────
nombre  = 'Maria Garcia Hernandez'
num_pol = 'GMM-2026-00145'
ramo    = 'Gastos Medicos Mayores'

# Longitud
print(f'Longitud del nombre: {len(nombre)} caracteres')

# Acceso por indice (Python empieza en 0)
print(f'Primera letra:  {nombre[0]}')     # M
print(f'Ultimas 3:      {nombre[-3:]}')

# Slicing: [inicio:fin] (fin excluido)
tipo_poliza = num_pol[0:3]    # 'GMM'
year_emision = num_pol[4:8]   # '2026'
print(f'Tipo poliza: {tipo_poliza}')
print(f'Year emision: {year_emision}')
````

````python
# ── Metodos mas usados en limpieza de datos ──────────────────────────────────
# Al leer CSVs de sistemas legacy, los datos tienen problemas de formato

nombre_sucio = '  JUAN PEREZ  '
print(f'strip():  |{nombre_sucio.strip()}|')
print(f'title():  {nombre_sucio.strip().title()}')
print(f'lower():  {nombre_sucio.strip().lower()}')

poliza = 'GMM-2026-00145'
print(f'upper():    {poliza.upper()}')
print(f'replace():  {poliza.replace("-", "_")}')
print(f'startswith: {poliza.startswith("GMM")}')
print(f'contains:   {"2026" in poliza}')

# Dividir y unir
partes = poliza.split('-')
print(f'split: {partes}')
print(f'join:  {"|" .join(partes)}')
````

### 7. bool y None — Logicos y Valores Ausentes

Los booleanos son la base de **todos los filtros** en pandas.
Entenderlos bien te ahorra horas de debugging.

````python
# ── bool: True / False ───────────────────────────────────────────────────────
poliza_activa   = True
tiene_siniestro = False

# Operadores de comparacion — devuelven bool
edad = 58
print(f'edad > 60:  {edad > 60}')    # False
print(f'edad >= 18: {edad >= 18}')   # True
print(f'edad == 58: {edad == 58}')   # True  (== es comparacion, = es asignacion)
print(f'edad != 30: {edad != 30}')   # True

# Operadores logicos: and / or / not
suma = 800_000
preexistencias = False
acepta = (edad <= 65) and (suma <= 1_000_000) and not preexistencias
print(f'Acepta automatico: {acepta}')   # True
````

````python
# ── Regla de aceptacion de polizas GMM ───────────────────────────────────────

def evaluar_poliza(edad, suma_asegurada, imc, preexistencias, siniestros_3a):
    # Condiciones de rechazo directo
    if edad > 70 or edad < 18:
        return 'DECLINA'
    if siniestros_3a >= 3:
        return 'DECLINA'
    # Condiciones de revision medica
    requiere_revision = (
        (edad >= 55 and suma_asegurada > 1_000_000) or
        (imc >= 35) or
        (preexistencias and suma_asegurada > 500_000)
    )
    return 'REVISION_MEDICA' if requiere_revision else 'ACEPTA'


perfiles = [
    ('Ana Torres',  28, 500_000, 22.5, False, 0),
    ('Luis Perez',  58, 2_000_000, 28.1, True,  1),
    ('Rosa Mendez', 35, 300_000, 37.2, False, 0),
    ('Jorge Silva', 72, 500_000, 25.0, False, 0),
]

print(f'{"Asegurado":<18} {"Edad":>5} {"Suma":>12} {"Resultado":<18}')
print('-' * 56)
for nombre, edad, suma, imc, preex, siniest in perfiles:
    r = evaluar_poliza(edad, suma, imc, preex, siniest)
    print(f'{nombre:<18} {edad:>5} ${suma:>10,} {r:<18}')
````

````python
# ── None: ausencia de valor ──────────────────────────────────────────────────
# Equivalente a NULL (SQL), NA (R)

beneficiario  = None    # poliza sin beneficiario registrado
monto_reserva = None    # siniestro sin valuar

# IMPORTANTE: usa 'is' para comparar con None, no '=='
print(f'beneficiario is None:     {beneficiario is None}')      # True
print(f'beneficiario is not None: {beneficiario is not None}')  # False

print(f'bool(None):  {bool(None)}')    # False
print(f'not None:    {not None}')       # True

# Patron comun: valor por defecto si es None
def obtener_deducible(valor, default=5000):
    return valor if valor is not None else default

print(f'Deducible no registrado: ${obtener_deducible(None):,}')
print(f'Deducible registrado:    ${obtener_deducible(8000):,}')
````

### 8. datetime — Fechas

Las fechas son criticas en seguros: vigencias, exposicion, dias de mora,
aging de siniestros, cortes de reservas.

````python
# ── Crear y leer fechas ──────────────────────────────────────────────────────
from datetime import date, datetime, timedelta

# Fecha de hoy
hoy = date.today()
print(f'Hoy: {hoy}')
print(f'Year: {hoy.year} | Mes: {hoy.month} | Dia: {hoy.day}')

# Fechas especificas
fecha_emision   = date(2026, 1, 15)
fecha_vence     = date(2026, 12, 31)
fecha_siniestro = date(2026, 3, 20)

print(f'Emision:     {fecha_emision}')
print(f'Vencimiento: {fecha_vence}')
print(f'Siniestro:   {fecha_siniestro}')
````

````python
# ── Aritmetica de fechas ─────────────────────────────────────────────────────

# Diferencia entre fechas
dias_vigencia = (fecha_vence - fecha_emision).days
print(f'Dias de vigencia: {dias_vigencia}')

# Exposicion transcurrida
dias_expuesto = (hoy - fecha_emision).days
fraccion_exp  = dias_expuesto / dias_vigencia
print(f'Dias expuestos:    {dias_expuesto}')
print(f'Fraccion expuesta: {fraccion_exp:.2%}')

# Dias de reporte del siniestro
dias_reporte = (hoy - fecha_siniestro).days
print(f'Dias desde siniestro: {dias_reporte}')

# Sumar / restar dias
renovacion     = fecha_vence + timedelta(days=1)
aviso_30_dias  = fecha_vence - timedelta(days=30)
print(f'Fecha renovacion:   {renovacion}')
print(f'Aviso vencimiento:  {aviso_30_dias}')
````

````python
# ── Calcular edad actuarial ───────────────────────────────────────────────────

def calcular_edad(fecha_nac, fecha_ref=None):
    if fecha_ref is None:
        fecha_ref = date.today()
    edad = fecha_ref.year - fecha_nac.year
    # Ajuste si aun no cumplio years este year
    if (fecha_ref.month, fecha_ref.day) < (fecha_nac.month, fecha_nac.day):
        edad -= 1
    return edad


asegurados = [
    ('Ana Torres',  date(1998, 6, 20)),
    ('Luis Perez',  date(1980, 3, 15)),
    ('Rosa Mendez', date(1965, 11, 30)),
]

print(f'{"Nombre":<18} {"Nacimiento":<14} {"Edad":>5}')
print('-' * 40)
for nombre, nac in asegurados:
    print(f'{nombre:<18} {nac}   {calcular_edad(nac):>5}')
````

````python
# ── Formatos de fecha para reportes (strftime) ───────────────────────────────
hoy = date.today()

print('Formatos comunes en seguros:')
print(f'ISO 8601 (BD):     {hoy.strftime("%Y-%m-%d")}')
print(f'Mexico:            {hoy.strftime("%d/%m/%Y")}')
print(f'Mes-Year:          {hoy.strftime("%m/%Y")}')
print(f'Year-Mes:          {hoy.strftime("%Y-%m")}')
print(f'Mes completo:      {hoy.strftime("%B %Y")}')

# strptime: string → date (muy util al leer CSVs)
fecha_str = '15/01/2026'
fecha_obj = datetime.strptime(fecha_str, '%d/%m/%Y').date()
print(f'Convertir {fecha_str!r} --> {fecha_obj}')

# Formato ISO — el mas comun en bases de datos
fecha_iso_obj = date.fromisoformat('2026-03-20')
print(f'ISO fromisoformat --> {fecha_iso_obj}')
````

**Contexto:** Recibes los datos de 5 polizas de seguro de vida y necesitas
generar un reporte de seguimiento. Este ejercicio usa **todos los tipos** vistos hoy.

````python
# ── Datos del ejercicio ──────────────────────────────────────────────────────
from datetime import date

polizas = [
    {'id': 'VID-2026-001', 'asegurado': 'sofia ramirez mendoza',
     'fecha_nac': date(1990, 3, 15), 'fecha_emision': date(2026, 2, 1),
     'fecha_vence': date(2027, 1, 31), 'suma_aseg': 500_000,
     'prima_anual': 3_200.0, 'activa': True, 'beneficiario': None},

    {'id': 'VID-2026-002', 'asegurado': '  MIGUEL TORRES LUNA  ',
     'fecha_nac': date(1975, 8, 22), 'fecha_emision': date(2026, 1, 15),
     'fecha_vence': date(2027, 1, 14), 'suma_aseg': 1_500_000,
     'prima_anual': 9_450.0, 'activa': True, 'beneficiario': 'Elena Torres'},

    {'id': 'VID-2026-003', 'asegurado': 'Laura Gonzalez',
     'fecha_nac': date(1955, 12, 5), 'fecha_emision': date(2025, 7, 1),
     'fecha_vence': date(2026, 6, 30), 'suma_aseg': 300_000,
     'prima_anual': 4_100.0, 'activa': False, 'beneficiario': 'Roberto Gonzalez'},

    {'id': 'VID-2026-004', 'asegurado': 'carlos ruiz vega',
     'fecha_nac': date(2001, 5, 18), 'fecha_emision': date(2026, 3, 10),
     'fecha_vence': date(2027, 3, 9), 'suma_aseg': 250_000,
     'prima_anual': 1_850.0, 'activa': True, 'beneficiario': None},

    {'id': 'VID-2026-005', 'asegurado': 'Patricia Nava',
     'fecha_nac': date(1968, 9, 30), 'fecha_emision': date(2026, 4, 1),
     'fecha_vence': date(2027, 3, 31), 'suma_aseg': 800_000,
     'prima_anual': 6_700.0, 'activa': True, 'beneficiario': 'Hector Nava'},
]
print(f'{len(polizas)} polizas cargadas')
````

````python
# ── Tarea 1: Limpiar nombres ─────────────────────────────────────────────────
# Algunos nombres tienen espacios extra o estan en minusculas.
# Normaliza todos a formato Title Case sin espacios extra.

print(f'{"ID":<16} {"Nombre original":<28} {"Nombre limpio"}')
print('-' * 68)
for p in polizas:
    limpio = p['asegurado'].strip().title()
    print(f'{p["id"]:<16} {p["asegurado"].strip():<28} {limpio}')
````

````python
# ── Tarea 2: Calcular edad y dias de vigencia ────────────────────────────────
hoy = date.today()

print(f'{"ID":<16} {"Nombre":<18} {"Edad":>5} {"Dias vig.":>10} {"Dias exp.":>10}')
print('-' * 64)

for p in polizas:
    nombre = p['asegurado'].strip().title().split()[0]
    # Edad
    edad = hoy.year - p['fecha_nac'].year
    if (hoy.month, hoy.day) < (p['fecha_nac'].month, p['fecha_nac'].day):
        edad -= 1
    # Vigencia
    dias_vig = (p['fecha_vence'] - p['fecha_emision']).days
    dias_exp = max(0, min((hoy - p['fecha_emision']).days, dias_vig))
    print(f'{p["id"]:<16} {nombre:<18} {edad:>5} {dias_vig:>10,} {dias_exp:>10,}')
````

````python
# ── Tarea 3: Resumen de cartera ──────────────────────────────────────────────
polizas_activas  = [p for p in polizas if p['activa']]
sin_beneficiario = [p for p in polizas if p['beneficiario'] is None]
prima_total      = sum(p['prima_anual'] for p in polizas)
prima_activas    = sum(p['prima_anual'] for p in polizas_activas)
suma_total       = sum(p['suma_aseg']   for p in polizas)

print('=' * 52)
print('  REPORTE DE CARTERA — VIDA INDIVIDUAL')
print(f'  Fecha de corte: {hoy.strftime("%d/%m/%Y")}')
print('=' * 52)
print(f'  Total polizas              : {len(polizas):>8}')
print(f'  Polizas activas            : {len(polizas_activas):>8}')
print(f'  Sin beneficiario (alerta)  : {len(sin_beneficiario):>8}')
print(f'  Prima total cartera        : ${prima_total:>10,.2f}')
print(f'  Prima cartera activa       : ${prima_activas:>10,.2f}')
print(f'  Prima promedio             : ${prima_total/len(polizas):>10,.2f}')
print(f'  Suma asegurada total       : ${suma_total:>10,.0f}')
print('=' * 52)

print()
print('  Polizas sin beneficiario (requieren gestion):')
for p in sin_beneficiario:
    print(f'   {p["id"]}  -  {p["asegurado"].strip().title()}')
````

````python
# ── Tarea 4 (Bonus): Avisos de vencimiento ───────────────────────────────────
dias_alerta = 90
print(f'Polizas que vencen en los proximos {dias_alerta} dias:')
print()

hay_alertas = False
for p in polizas:
    dias_para_vencer = (p['fecha_vence'] - hoy).days
    if 0 <= dias_para_vencer <= dias_alerta:
        hay_alertas = True
        nombre = p['asegurado'].strip().title()
        estado = 'ACTIVA' if p['activa'] else 'INACTIVA'
        print(f'  {p["id"]}')
        print(f'     Asegurado:   {nombre}')
        print(f'     Vencimiento: {p["fecha_vence"].strftime("%d/%m/%Y")} ({dias_para_vencer} dias)')
        print(f'     Prima:       ${p["prima_anual"]:,.2f}/anio')
        print(f'     Estado:      {estado}')
        print()

if not hay_alertas:
    print('  No hay polizas por vencer en los proximos 90 dias')
````
