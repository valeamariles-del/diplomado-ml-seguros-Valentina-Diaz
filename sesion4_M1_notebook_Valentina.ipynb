## 1. El Problema: Codigo que se Repite

Antes de aprender la solucion, necesitamos sentir el dolor del problema.
Ejecuta la siguiente celda y observa cuantas veces repites exactamente lo mismo:

````python
# ── Sin funciones: calculo de prima para 4 polizas ──────────────────────────
# Observa que el calculo es IDENTICO en los 4 casos — solo cambian los valores

# Poliza 1
suma1, tasa1 = 500_000, 0.022
prima1 = suma1 * tasa1 * (1 + 0.15) * 1.16
print(f'Poliza 1: ${prima1:,.2f}')

# Poliza 2
suma2, tasa2 = 1_000_000, 0.031
prima2 = suma2 * tasa2 * (1 + 0.15) * 1.16
print(f'Poliza 2: ${prima2:,.2f}')

# Poliza 3
suma3, tasa3 = 300_000, 0.018
prima3 = suma3 * tasa3 * (1 + 0.15) * 1.16
print(f'Poliza 3: ${prima3:,.2f}')

# Poliza 4
suma4, tasa4 = 2_000_000, 0.048
prima4 = suma4 * tasa4 * (1 + 0.15) * 1.16
print(f'Poliza 4: ${prima4:,.2f}')
````

**Pregunta:** Si manana la SHCP sube el IVA del 16% al 17%, cuantas lineas
tendrias que cambiar en el codigo anterior?

**Respuesta:** 4 lineas. Y si la cartera tiene 10,000 polizas, serian 10,000 lineas.
Ademas, si te equivocas en una, el error solo afecta a esa poliza
y puede pasar desapercibido.

> **Principio DRY (Don't Repeat Yourself):** Si escribes el mismo codigo
> mas de una vez, es candidato a convertirse en una funcion.

<a id='2-solucion'></a>
## 2. La Solucion: Funciones

Una **funcion** es un bloque de codigo con nombre que encapsula una tarea.
Lo escribes una vez y lo reutilizas tantas veces como necesites.

````python
# ── Con funcion: el mismo calculo para 4 polizas ────────────────────────────
# El calculo esta en UN solo lugar

def calcular_prima(suma, tasa):
    prima = suma * tasa * (1 + 0.15) * 1.16
    return prima

# Usar la funcion es trivial
print(f'Poliza 1: ${calcular_prima(500_000,   0.022):,.2f}')
print(f'Poliza 2: ${calcular_prima(1_000_000, 0.031):,.2f}')
print(f'Poliza 3: ${calcular_prima(300_000,   0.018):,.2f}')
print(f'Poliza 4: ${calcular_prima(2_000_000, 0.048):,.2f}')

# Si sube el IVA: cambias en 1 sola linea y afecta a las 4 polizas
````

## 3. Anatomia de una Funcion

```
def calcular_prima(suma, tasa):
    # cuerpo — va indentado 4 espacios
    resultado = suma * tasa
    return resultado
```

| Parte | Que es |
|-------|--------|
| `def` | Palabra reservada que inicia TODA funcion |
| `calcular_prima` | El nombre que tu eliges (snake_case) |
| `(suma, tasa)` | Los parametros: datos que recibe la funcion |
| `:` | Dos puntos obligatorios — inician el bloque |
| Cuerpo indentado | Las instrucciones de la funcion (4 espacios) |
| `return` | Devuelve el resultado al lugar donde se llamo |

> **Definir NO es ejecutar.** Cuando escribes `def calcular_prima(...):`
> Python registra la funcion en memoria pero NO la ejecuta.
> Para ejecutarla debes **llamarla**: `calcular_prima(500_000, 0.022)`

````python
# Demostración: definir y luego llamar son dos pasos distintos

# Paso 1: DEFINIR la funcion (Python la registra en memoria, no la ejecuta)
def saludar_diplomado():
    print('Bienvenidos al Diplomado ML en Seguros')
    print('Modulo 1 - Sesion 4')

# En este punto la funcion EXISTE pero no ha hecho nada todavia
print('Funcion definida. Tipo:', type(saludar_diplomado))

# Paso 2: LLAMAR la funcion — ahora si se ejecuta
saludar_diplomado()

# Puedes llamarla las veces que quieras
saludar_diplomado()
````

## 4. Progresion: De Simple a Compleja

Las funciones se construyen gradualmente.
Empecemos desde la mas basica y vamos agregando capacidades.

````python
# ── Nivel 1: Sin parametros ──────────────────────────────────────────────────
# La funcion mas sencilla posible — hace siempre lo mismo

def mostrar_encabezado():
    print('=' * 45)
    print('  REPORTE DE CARTERA — GMM INDIVIDUAL')
    print('=' * 45)

mostrar_encabezado()
print('  ... contenido del reporte ...')
mostrar_encabezado()   # reutilizamos para el pie
````

````python
# ── Nivel 2: Con un parametro ───────────────────────────────────────────────
# Ahora la funcion recibe un dato y lo usa

def mostrar_prima(prima):
    print(f'Prima calculada: ${prima:,.2f} MXN')

mostrar_prima(13_754.0)
mostrar_prima(28_710.4)
mostrar_prima(4_089.6)
````

````python
# ── Nivel 3: Con varios parametros y return ─────────────────────────────────
# La funcion calcula algo y lo devuelve para poder usarlo

def prima_pura(suma_asegurada, tasa_riesgo):
    return suma_asegurada * tasa_riesgo

# El resultado se puede guardar y usar
p1 = prima_pura(500_000, 0.022)
p2 = prima_pura(1_000_000, 0.031)

print(f'Prima pura p1: ${p1:,.2f}')
print(f'Prima pura p2: ${p2:,.2f}')
print(f'Prima total:   ${p1 + p2:,.2f}')   # podemos operar con el resultado
````

````python
# ── Nivel 4: Con parametros por defecto ─────────────────────────────────────
# Algunos parametros tienen un valor 'normal' que aplica la mayoria de las veces

def calcular_prima(suma, tasa, recargo=0.15, iva=1.16):
    return suma * tasa * (1 + recargo) * iva

# Llamadas con distintas combinaciones:
p_normal  = calcular_prima(500_000, 0.022)                # usa defaults
p_alto    = calcular_prima(500_000, 0.022, recargo=0.25)  # recargo mayor
p_sin_iva = calcular_prima(500_000, 0.022, iva=1.0)       # sin IVA

print(f'Prima normal:     ${p_normal:,.2f}')
print(f'Prima alto riesgo:${p_alto:,.2f}')
print(f'Prima sin IVA:    ${p_sin_iva:,.2f}')
````

## 5. Parametros a Fondo

Los parametros son los 'datos de entrada' de la funcion.
Hay tres tipos principales:

| Tipo | Descripcion | Ejemplo |
|------|-------------|---------|
| Obligatorio | Siempre debe pasarse | `def f(suma, tasa):` |
| Por defecto | Tiene valor si no se pasa | `def f(suma, tasa, iva=1.16):` |
| Keyword arg | Se pasa por nombre | `f(suma=500_000, tasa=0.022)` |

````python
# ── Regla: parametros por defecto SIEMPRE al final ──────────────────────────

# CORRECTO: obligatorios primero, defaults despues
def emitir_poliza(nombre, suma, ramo, activa=True, deducible=5000):
    estado = 'ACTIVA' if activa else 'INACTIVA'
    return f'{ramo} | {nombre} | ${suma:,} | Deduc: ${deducible:,} | {estado}'

# Distintas formas de llamar la misma funcion:
print(emitir_poliza('Ana Torres', 500_000, 'GMM'))              # defaults
print(emitir_poliza('Luis Perez', 1_000_000, 'GMM', False))     # activa=False
print(emitir_poliza('Maria G.', 300_000, 'GMM', deducible=10_000)) # solo deducible

# Con nombres: el orden no importa
print(emitir_poliza(ramo='Autos', nombre='Carlos R.', suma=800_000))
````

````python
# ── TypeError: que pasa si no pasas un argumento obligatorio ────────────────

def calcular_prima(suma, tasa):
    return suma * tasa * 1.15 * 1.16

# Llamada correcta
print(f'${calcular_prima(500_000, 0.022):,.2f}')

# Llamada incorrecta — Python te dice exactamente que falta
try:
    calcular_prima(500_000)   # falta tasa
except TypeError as e:
    print(f'Error: {e}')
    print('Debes pasar todos los argumentos obligatorios.')
````

## 6. return: Devolver Resultados

La diferencia entre `print` y `return` es fundamental y confunde mucho al principio:

- **`print`** muestra algo en pantalla — el valor se pierde, no puedes usarlo
- **`return`** produce un valor que puedes guardar, operar o pasar a otra funcion

````python
# ── print vs return — la diferencia en la practica ─────────────────────────

def prima_con_print(suma, tasa):
    resultado = suma * tasa * 1.16
    print(resultado)   # solo muestra, no produce

def prima_con_return(suma, tasa):
    resultado = suma * tasa * 1.16
    return resultado   # produce el valor

# Con print: el resultado se pierde
p_print = prima_con_print(500_000, 0.022)
print('Tipo:', type(p_print), '| Valor:', p_print)   # None !

# Con return: el resultado se puede usar
p_return = prima_con_return(500_000, 0.022)
print('Tipo:', type(p_return), '| Valor:', p_return)
print('Mensual:', p_return / 12)   # podemos operar con el
````

````python
# ── Devolver multiples valores ───────────────────────────────────────────────
# Python permite devolver varios valores separados por coma (los empaca en tupla)

def analizar_cartera(primas):
    total    = sum(primas)
    promedio = total / len(primas)
    maximo   = max(primas)
    minimo   = min(primas)
    return total, promedio, maximo, minimo   # devuelve 4 valores

primas_cartera = [2400, 6200, 1900, 14500, 3200, 8800]

# Desempaquetar en variables separadas
total, prom, maximo, minimo = analizar_cartera(primas_cartera)
print(f'Total:   ${total:>10,.2f}')
print(f'Promedio:${prom:>10,.2f}')
print(f'Maximo:  ${maximo:>10,.2f}')
print(f'Minimo:  ${minimo:>10,.2f}')
````

````python
# ── return con condicion: salida anticipada ──────────────────────────────────
# Python sale de la funcion en cuanto encuentra un return

def clasificar_edad_asegurable(edad):
    if edad < 18:
        return 'MENOR — no asegurable'      # sale aqui si edad < 18
    if edad > 70:
        return 'MAYOR 70 — requiere estudio' # sale aqui si edad > 70
    if edad <= 30:
        return 'GRUPO A (18-30)'
    if edad <= 50:
        return 'GRUPO B (31-50)'
    return 'GRUPO C (51-70)'                 # solo llega aqui si ninguna anterior

edades = [15, 25, 38, 62, 72, 45]
for edad in edades:
    print(f'Edad {edad:>3}: {clasificar_edad_asegurable(edad)}')
````

## 7. Scope: Ambito de Variables

El **scope** define donde existe y es accesible una variable.
Variables creadas DENTRO de una funcion son **locales** —
solo existen mientras la funcion se ejecuta.

````python
# ── Variable local: solo vive dentro de la funcion ─────────────────────────

def calcular_prima(suma, tasa):
    prima_neta  = suma * tasa          # variable LOCAL
    prima_total = prima_neta * 1.16    # variable LOCAL
    return prima_total

resultado = calcular_prima(500_000, 0.022)
print(f'Resultado: ${resultado:,.2f}')   # OK — resultado es GLOBAL

# Las variables locales NO existen fuera:
try:
    print(prima_neta)   # NameError
except NameError as e:
    print(f'Error esperado: {e}')
    print('prima_neta solo existe dentro de calcular_prima()')
````

````python
# ── Variable global: constantes del programa ─────────────────────────────────
# Usa variables globales solo para constantes que no cambian

# Buena practica: constantes en MAYUSCULAS (convencion PEP8)
IVA_MEXICO   = 1.16
RECARGO_BASE = 0.15

def calcular_prima(suma, tasa):
    # Lee las constantes globales (no las modifica)
    prima_neta = suma * tasa * (1 + RECARGO_BASE)
    return prima_neta * IVA_MEXICO

print(f'${calcular_prima(500_000, 0.022):,.2f}')

# Mejor practica aun: pasar como parametro con default
def calcular_prima_v2(suma, tasa, recargo=RECARGO_BASE, iva=IVA_MEXICO):
    return suma * tasa * (1 + recargo) * iva

print(f'${calcular_prima_v2(500_000, 0.022):,.2f}')   # identico
print(f'${calcular_prima_v2(500_000, 0.022, iva=1.08):,.2f}')  # facil de cambiar
````

## 8. Docstrings y PEP8

**Docstring:** comentario especial entre triple comillas inmediatamente
despues de `def`. Documenta que hace la funcion, sus parametros y lo que devuelve.

**PEP8:** guia oficial de estilo de Python. Las convenciones mas importantes:

| Convencion | Ejemplo correcto | Ejemplo incorrecto |
|-----------|-----------------|-------------------|
| Nombres en snake_case | `calcular_prima` | `calcularPrima` |
| Constantes en MAYUSCULAS | `IVA = 1.16` | `iva = 1.16` |
| 4 espacios de indentacion | `    return x` | `  return x` |
| Espacios alrededor de `=` | `prima = suma * tasa` | `prima=suma*tasa` |
| Lineas de max 79 caracteres | (particionar si es muy larga) | |

````python
# ── Funcion con docstring completo ───────────────────────────────────────────
# Este es el estandar que usaremos en el diplomado

def calcular_prima(suma_asegurada, tasa_riesgo, recargo=0.15, iva=1.16):
    """
    Calcula la prima comercial de una poliza de seguros.
    
    Parametros
    ----------
    suma_asegurada : float
        Suma asegurada en pesos MXN.
    tasa_riesgo : float
        Tasa tecnica de riesgo (ej. 0.022 para 2.2%).
    recargo : float, optional
        Recargo de seguridad, por defecto 0.15 (15%).
    iva : float, optional
        Factor de IVA, por defecto 1.16 (16%).
    
    Retorna
    -------
    float
        Prima comercial con IVA incluido en MXN.
    
    Ejemplo
    -------
    >>> calcular_prima(500_000, 0.022)
    13754.0
    """
    prima_pura      = suma_asegurada * tasa_riesgo
    prima_comercial = prima_pura * (1 + recargo)
    return prima_comercial * iva

# Verificar que el docstring funciona
help(calcular_prima)
````

````python
# ── Usar la funcion documentada ──────────────────────────────────────────────
prima = calcular_prima(500_000, 0.022)
print(f'Prima: ${prima:,.2f}')

# El docstring aparece en el tooltip de VS Code y Jupyter
# cuando escribes el nombre de la funcion
# calcular_prima?   # esto funciona en Jupyter directamente
````
